```python
import math
import random
from dataclasses import dataclass
from typing import Dict, Tuple

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Helpers: Sharpe & utilities
# -----------------------------

def compute_daily_returns(prices: pd.Series) -> pd.Series:
    # Calculate percentage change and fill NaN values (first row) with 0
    returns = prices.pct_change().fillna(0.0)
    # Replace infinite values (resulting from division by zero due to padding) with 0
    returns = returns.replace([np.inf, -np.inf], 0.0)
    return returns

def annualized_sharpe(returns: pd.Series, risk_free_daily: float = 0.0, trading_days: int = 252) -> float:
    excess = returns - risk_free_daily
    mu = excess.mean()
    sigma = excess.std(ddof=0)
    if sigma == 0:
        return 0.0
    return (mu / sigma) * math.sqrt(trading_days)

def equity_curve_from_returns(returns: pd.Series, start_value: float = 1.0) -> pd.Series:
    return start_value * (1.0 + returns).cumprod()

# -----------------------------
# Simple SMA Strategy (baseline)
# -----------------------------

def sma_strategy(prices: pd.Series, fast: int = 20, slow: int = 50) -> pd.Series:
    fast_ma = prices.rolling(fast).mean()
    slow_ma = prices.rolling(slow).mean()
    signal = np.sign(fast_ma - slow_ma).fillna(0.0)
    rets = compute_daily_returns(prices)
    strat_rets = signal.shift(1).fillna(0.0) * rets
    strat_rets.name = "SMA_returns"
    return strat_rets

# ----------------------------------------
# RL Environment (single asset, discrete)
# ----------------------------------------

@dataclass
class EnvConfig:
    window: int = 30
    trans_cost: float = 0.0002
    risk_aversion: float = 5.0
    max_position: int = 1
    trading_days: int = 252

class SingleAssetTradingEnv:
    def __init__(self, prices: pd.Series, cfg: EnvConfig):
        self.prices = prices.dropna().copy()
        # Ensure returns is a Series for compute_daily_returns
        if isinstance(self.prices, pd.DataFrame):
             self.returns = compute_daily_returns(self.prices.iloc[:, 0]).values # Take the first column if DataFrame
        else:
             self.returns = compute_daily_returns(self.prices).values
        self.cfg = cfg
        self.T = len(self.returns)
        self.reset()
        self.num_equities = self.prices.shape[0] if isinstance(self.prices, pd.DataFrame) else 1
        self.corr_matrix = np.zeros((self.num_equities,self.num_equities))


    def reset(self) -> np.ndarray:
        self.t = self.cfg.window
        self.position = 0
        if self.T <= self.cfg.window: # Add check for insufficient data
            return None
        return self._get_state()

    def compute_correlation_matrix(self) -> np.ndarray:
      self.corr_matrix = np.corrcoef(self.returns.T)
      print(self.corr_matrix.shape)
      return self.corr_matrix

    def compute_portifolio_variance(self, weights: np.ndarray) -> float:
        return np.std(self.returns)
        # return np.dot(weights, np.dot(self.corr_matrix, weights))

    def _get_state(self) -> np.ndarray:
        window_ret = self.returns[self.t - self.cfg.window:self.t]
        pos_oh = np.zeros(3)
        pos_oh[self.position + 1] = 1.0
        return np.concatenate([window_ret.flatten(), pos_oh]) # Flatten window_ret to ensure 1D


    def step(self, action: int) -> Tuple[np.ndarray, float, bool, dict]:
        new_position = action - 1
        new_position = int(np.clip(new_position, -self.cfg.max_position, self.cfg.max_position))
        r_t = self.returns[self.t]
        pnl = self.position * r_t # profit and loss
        cost = self.cfg.trans_cost * abs(new_position - self.position)
        volatility = self.compute_portifolio_variance(new_position)
        reward = pnl - cost - self.cfg.risk_aversion * (volatility ** 2)

        self.position = new_position
        self.t += 1
        done = self.t >= self.T - 1
        state = self._get_state() if not done else None
        info = {"pnl": pnl, "cost": cost, "raw_ret": r_t, "position": self.position}
        return state, float(reward), bool(done), info

# ----------------------------------------
# DQN (NumPy) with Experience Replay
# ----------------------------------------

@dataclass
class DQNConfig:
    state_dim: int
    action_dim: int = 3
    hidden_dim: int = 64
    gamma: float = 0.99
    lr: float = 1e-3
    batch_size: int = 64
    buffer_size: int = 20000
    min_buffer: int = 512
    epsilon_start: float = 1.0
    epsilon_end: float = 0.05
    epsilon_decay_steps: int = 5000
    target_update_every: int = 200
    train_steps: int = 15000
    seed: int = 42

class ReplayBuffer:
    def __init__(self, capacity: int, state_dim: int):
        self.capacity = capacity
        self.s = np.zeros((capacity, state_dim), dtype=np.float32)
        self.a = np.zeros((capacity,), dtype=np.int64)
        self.r = np.zeros((capacity,), dtype=np.float32)
        self.ns = np.zeros((capacity, state_dim), dtype=np.float32)
        self.d = np.zeros((capacity,), dtype=np.float32)
        self.ptr = 0
        self.size = 0

    def add(self, s, a, r, ns, d):
        self.s[self.ptr] = s
        self.a[self.ptr] = a
        self.r[self.ptr] = r
        self.ns[self.ptr] = ns if ns is not None else 0.0
        self.d[self.ptr] = d
        self.ptr = (self.ptr + 1) % self.capacity
        self.size = min(self.size + 1, self.capacity)

    def sample(self, batch_size):
        idx = np.random.randint(0, self.size, size=batch_size)
        return (self.s[idx], self.a[idx], self.r[idx], self.ns[idx], self.d[idx])

def xavier_init(in_dim, out_dim):
    limit = math.sqrt(6 / (in_dim + out_dim))
    return np.random.uniform(-limit, limit, (in_dim, out_dim)).astype(np.float32), np.zeros((out_dim,), dtype=np.float32)

class Dense:
    def __init__(self, in_dim, out_dim):
        W, b = xavier_init(in_dim, out_dim)
        self.W = W
        self.b = b
        self.dW = np.zeros_like(W)
        self.db = np.zeros_like(b)

    def forward(self, x):
        self.x = x
        return x @ self.W + self.b

    def backward(self, grad_out):
        self.dW = self.x.T @ grad_out
        self.db = grad_out.sum(axis=0)
        return grad_out @ self.W.T

    def step(self, lr):
        self.W -= lr * self.dW
        self.b -= lr * self.db

class ReLU:
    def forward(self, x):
        self.mask = (x > 0).astype(np.float32)
        return x * self.mask
    def backward(self, grad_out):
        return grad_out * self.mask

class QNetwork:
    def __init__(self, state_dim, hidden_dim, action_dim):
        self.l1 = Dense(state_dim, hidden_dim)
        self.act1 = ReLU()
        self.l2 = Dense(hidden_dim, hidden_dim)
        self.act2 = ReLU()
        self.out = Dense(hidden_dim, action_dim)

    def forward(self, x):
        z1 = self.l1.forward(x)
        a1 = self.act1.forward(z1)
        z2 = self.l2.forward(a1)
        a2 = self.act2.forward(z2)
        q = self.out.forward(a2)
        return q

    def backward(self, grad_q):
        grad = self.out.backward(grad_q)
        grad = self.act2.backward(grad)
        grad = self.l2.backward(grad)
        grad = self.act1.backward(grad)
        grad = self.l1.backward(grad)

    def step(self, lr):
        for layer in [self.l1, self.l2, self.out]:
            layer.step(lr)

    def copy_from(self, other: "QNetwork"):
        for (s, o) in [(self.l1, other.l1), (self.l2, other.l2), (self.out, other.out)]:
            s.W[:] = o.W
            s.b[:] = o.b

class DQNAgent:
    def __init__(self, cfg: DQNConfig):
        np.random.seed(cfg.seed)
        random.seed(cfg.seed)
        self.cfg = cfg
        self.q = QNetwork(cfg.state_dim, cfg.hidden_dim, cfg.action_dim)
        self.target = QNetwork(cfg.state_dim, cfg.hidden_dim, cfg.action_dim)
        self.target.copy_from(self.q)
        self.buffer = ReplayBuffer(cfg.buffer_size, cfg.state_dim)
        self.train_step = 0

    def epsilon(self):
        t = min(self.train_step, self.cfg.epsilon_decay_steps)
        frac = 1.0 - t / self.cfg.epsilon_decay_steps
        return self.cfg.epsilon_end + (self.cfg.epsilon_start - self.cfg.epsilon_end) * max(frac, 0.0)

    def act(self, state: np.ndarray) -> int:
        if np.random.rand() < self.epsilon():
            return np.random.randint(self.cfg.action_dim)
        q = self.q.forward(state[None, :])
        return int(np.argmax(q[0]))

    def learn(self):
        if self.buffer.size < self.cfg.min_buffer:
            return
        s, a, r, ns, d = self.buffer.sample(self.cfg.batch_size)
        q = self.q.forward(s)
        q_a = q[np.arange(self.cfg.batch_size), a]

        next_q = self.q.forward(ns)
        next_actions = np.argmax(next_q, axis=1)
        target_q = self.target.forward(ns)
        target_next = target_q[np.arange(self.cfg.batch_size), next_actions]

        y = r + (1 - d) * self.cfg.gamma * target_next

        grad_q = np.zeros_like(q)
        grad_q[np.arange(self.cfg.batch_size), a] = (q_a - y) / self.cfg.batch_size

        self.q.backward(grad_q)
        self.q.step(self.cfg.lr)

        self.train_step += 1
        if self.train_step % self.cfg.target_update_every == 0:
            self.target.copy_from(self.q)

    def fit_env(self, env: SingleAssetTradingEnv):
        s = env.reset()
        if s is None: # Check if reset returned None due to insufficient data
            print(f"Skipping training for {env.prices.name if isinstance(env.prices, pd.Series) else 'this asset'} due to insufficient data.")
            return False # Indicate that training was skipped

        for _ in range(self.cfg.min_buffer):
            a = np.random.randint(self.cfg.action_dim)
            ns, r, done, _ = env.step(a)
            self.buffer.add(s, a, r, ns if not done else s, float(done))
            s = env.reset() if done else ns

        s = env.reset()
        if s is None: # Check again after filling buffer
             print(f"Skipping training for {env.prices.name if isinstance(env.prices, pd.Series) else 'this asset'} due to insufficient data.")
             return False # Indicate that training was skipped

        for _ in range(self.cfg.train_steps):
            a = self.act(s)
            ns, r, done, info = env.step(a)
            self.buffer.add(s, a, r, ns if not done else s, float(done))
            self.learn()
            s = env.reset() if done else ns
        return True # Indicate that training was successful

# -----------------------------
# Backtest Utilities
# -----------------------------

def equal_weight_portfolio(returns_dict: Dict[str, pd.Series]) -> pd.Series:
    aligned = pd.concat(returns_dict.values(), axis=1).dropna(how="all")
    aligned = aligned.fillna(0.0)
    ew = aligned.mean(axis=1)
    ew.name = "Portfolio_returns"
    return ew

# -----------------------------
# CONFIG + Example Data
# -----------------------------

WINDOW = 30
ENV_CFG = EnvConfig(window=WINDOW, trans_cost=0.0002, risk_aversion=5.0, max_position=1)

DQN_BASE = {
    "action_dim": 3,
    "hidden_dim": 64,
    "gamma": 0.99,
    "lr": 1e-3,
    "batch_size": 64,
    "buffer_size": 20000,
    "min_buffer": 512,
    "epsilon_start": 1.0,
    "epsilon_end": 0.05,
    "epsilon_decay_steps": 5000,
    "target_update_every": 200,
    "train_steps": 6000,
    "seed": 42,
}

# Example synthetic data
np.random.seed(7)
dates = pd.bdate_range("2020-01-01", "2023-12-29")
def make_price(mu=0.0003, sigma=0.02, start=100.0):
    rets = np.random.normal(mu, sigma, len(dates))
    price = start * np.cumprod(1 + rets)
    return pd.Series(price, index=dates)

price_series_dict = {
    "ASSET_A": make_price(mu=0.0004, sigma=0.018, start=100),
    "ASSET_B": make_price(mu=0.0002, sigma=0.022, start=50),
    "ASSET_C": make_price(mu=0.00025, sigma=0.019, start=200),
}

price_series_dict = equity_data # Use the fetched equity data


split_date = "2025-01-01"
train_prices = {k: v.loc[:split_date] for k, v in price_series_dict.items()}
test_prices  = {k: v.loc[split_date:] for k, v in price_series_dict.items()}

# ----------------------------------------
# Train + Backtest
# ----------------------------------------

rl_returns_dict = {}
sma_returns_dict = {}

for ticker, px in train_prices.items():
    # Add a check for empty DataFrame before proceeding
    if px.empty:
        print(f"Skipping training and testing for {ticker} due to empty training data.")
        continue

    state_dim = ENV_CFG.window + 3
    dqn_cfg = DQNConfig(state_dim=state_dim, **DQN_BASE)

    env_train = SingleAssetTradingEnv(px, ENV_CFG)
    agent = DQNAgent(dqn_cfg)
    if agent.fit_env(env_train): # Check if training was successful
      env_test = SingleAssetTradingEnv(test_prices[ticker], ENV_CFG)
      s = env_test.reset()
      pnl_series = []
      # Add check for insufficient data in test environment
      if s is not None:
        for t in range(ENV_CFG.window, len(env_test.returns) - 1):
            a = agent.act(s)
            ns, r, done, info = env_test.step(a)
            pnl_series.append(info["pnl"])
            if done: break
            s = ns

        idx = test_prices[ticker].index[ENV_CFG.window:ENV_CFG.window + len(pnl_series)]
        rl_returns_dict[ticker] = pd.Series(pnl_series, index=idx, name=f"{ticker}_RL")

        sma_ret = sma_strategy(test_prices[ticker].iloc[:, 0] if isinstance(test_prices[ticker], pd.DataFrame) else test_prices[ticker], fast=20, slow=50).loc[idx] # Handle potential DataFrame input
        sma_returns_dict[ticker] = sma_ret.rename(f"{ticker}_SMA")
      else:
          print(f"Skipping testing for {ticker} due to insufficient data in test environment.")


rl_portfolio = equal_weight_portfolio(rl_returns_dict)
sma_portfolio = equal_weight_portfolio(sma_returns_dict)

rl_sharpe = annualized_sharpe(rl_portfolio)
sma_sharpe = annualized_sharpe(sma_portfolio)

rl_equity = equity_curve_from_returns(rl_portfolio, 1.0)
sma_equity = equity_curve_from_returns(sma_portfolio, 1.0)

print("=== Portfolio Performance (Out-of-sample) ===")
print(f"RL (DQN) Annualized Sharpe: {rl_sharpe:.3f}")
print(f"SMA (20/50) Annualized Sharpe: {sma_sharpe:.3f}")

plt.figure(figsize=(10, 5))
plt.plot(rl_equity.index, rl_equity.values, label="RL (DQN)")
plt.plot(sma_equity.index, sma_equity.values, label="SMA 20/50")
plt.title("Equity Curve (Equal-Weight Portfolio)")
plt.legend()
plt.savefig("exemplo.png")
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# ===== Environment =====
class PortfolioEnv:
    def __init__(self, returns, window=20, cost=0.001):
        self.returns = returns
        self.n_assets = returns.shape[1]
        self.window = window
        self.cost = cost
        self.reset()

    def reset(self):
        self.t = self.window
        self.done = False
        self.weights = np.ones(self.n_assets) / self.n_assets
        self.episode_returns = []
        return self._get_state()

    def step(self, new_weights):
        old_weights = self.weights
        new_weights = np.clip(new_weights, 0, 1)
        new_weights /= (new_weights.sum() + 1e-8)  # normalize

        # turnover cost
        turnover = np.abs(new_weights - old_weights).sum()
        cost = self.cost * turnover

        # portfolio return
        r = self.returns.iloc[self.t].values
        port_ret = np.dot(new_weights, r) - cost

        self.weights = new_weights
        self.episode_returns.append(port_ret)

        self.t += 1
        if self.t >= len(self.returns):
            self.done = True

        return self._get_state(), port_ret, self.done

    def _get_state(self):
        return self.returns.iloc[self.t - self.window:self.t].values


# ===== Policy Network =====
class PolicyNet(nn.Module):
    def __init__(self, n_assets, window):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(n_assets * window, 128),
            nn.ReLU(),
            nn.Linear(128, n_assets),
            nn.Softmax(dim=-1)  # long-only portfolio weights
        )

    def forward(self, x):
        return self.fc(x)


# ===== Training with Sharpe reward =====
def sharpe_ratio(rets):
    rets = np.array(rets)
    if rets.std() == 0:
        return 0
    return np.sqrt(252) * rets.mean() / rets.std()

def train(env, episodes=200, lr=1e-3):
    policy = PolicyNet(env.n_assets, env.window)
    optimizer = optim.Adam(policy.parameters(), lr=lr)

    for ep in range(episodes):
        log_probs = []
        state = env.reset()
        done = False

        while not done:
            state_t = torch.tensor(state, dtype=torch.float32).unsqueeze(0)
            weights = policy(state_t).squeeze()

            # sample discrete asset for simplicity
            dist = torch.distributions.Categorical(probs=weights)
            action = dist.sample()
            one_hot = torch.nn.functional.one_hot(action, env.n_assets).float().numpy()

            log_probs.append(dist.log_prob(action))
            state, _, done = env.step(one_hot)

        # compute Sharpe of trajectory
        sr = sharpe_ratio(env.episode_returns)

        # REINFORCE update
        loss = -torch.stack(log_probs).sum() * sr
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if (ep+1) % 20 == 0:
            print(f"Episode {ep+1}, Sharpe: {sr:.3f}")

    return policy


# ===== Backtest =====
def backtest(env, policy):
    state = env.reset()
    done = False
    port_curve = [1.0]

    while not done:
        state_t = torch.tensor(state, dtype=torch.float32).unsqueeze(0)
        weights = policy(state_t).squeeze().detach().numpy()
        state, r, done = env.step(weights)
        port_curve.append(port_curve[-1] * (1 + r))

    port_curve = pd.Series(port_curve, index=env.returns.index[env.window-1:])
    return port_curve

# Fetch data
import yfinance as yf
tickers = [
    "AAPL", "MSFT", "GOOG", "AMZN", "FB", "TSLA", "BRK-A", "JPM", "JNJ", "V",
    "PG", "UNH", "HD", "MA", "VZ", "NVDA", "PYPL", "DIS", "ADBE", "NFLX",
    "CMCSA", "CSCO", "PEP", "KO", "MRK", "PFE", "TM", "BABA", "XOM", "WMT"
]
data = yf.download(tickers, period="max", group_by='column')  # max history
close = data['Close']
returns_temp = close.pct_change()
returns = returns_temp.dropna()
print(returns.shape,returns_temp.shape)

# Train RL agent
env = PortfolioEnv(returns, window=20, cost=0.001)
policy = train(env, episodes=300, lr=1e-3)

# Backtest RL
env_bt = PortfolioEnv(returns, window=20, cost=0.001)
rl_curve = backtest(env_bt, policy)
rl_length = len(rl_curve)

# Baseline: Equal Weight
ew_curve = (returns.iloc[-rl_length:].mean(axis=1) + 1).cumprod()

# Baseline: SMA on one asset (e.g., AAPL)
close_appl = close['AAPL'].dropna()
sma_fast = close_appl.rolling(20).mean()
sma_slow = close_appl.rolling(50).mean()
signal = (sma_fast > sma_slow).astype(int).shift(1).iloc[-rl_length:]
sma_curve = (1 + signal * returns["AAPL"].iloc[-rl_length:]).cumprod()

# Plot comparison
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(rl_curve, label="RL (Sharpe-Optimized)")
plt.plot(ew_curve, label="Equal Weight")
plt.plot(sma_curve, label="SMA (AAPL)")
plt.legend()
plt.title("Backtest: RL vs Baselines")
plt.show()

# Performance stats
def perf_stats(curve):
    rets = curve.pct_change().dropna()
    return {
        "Sharpe": sharpe_ratio(rets),
        "CAGR": (curve.iloc[-1]/curve.iloc[0])**(252/len(curve)) - 1,
        "MaxDD": (curve/curve.cummax() - 1).min()
    }

print("RL", perf_stats(rl_curve))
print("EW", perf_stats(ew_curve))
print("SMA", perf_stats(sma_curve))
```
